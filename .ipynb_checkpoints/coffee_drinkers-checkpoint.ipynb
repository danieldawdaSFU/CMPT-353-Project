```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from scipy import stats
import numpy as np
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
def coffees_num(d):
    if d == 'Less than 1':
        ret = 0
    elif d == 'More than 4':
        ret = 5
    else: 
        ret = int(d)

    return ret
```

# ANOVA Conditions and Test

## Ages

```python
df = pd.read_csv('data/GACTT_RESULTS_ANONYMIZED_v2.csv', index_col='Submission ID')

# we care about how much coffee people are drinking per day
num_drinks = df.dropna(subset=['How many cups of coffee do you typically drink per day?', 'Where do you typically drink coffee?'])

# we only really want people who drink coffee on the go or at cafes
num_drinks = num_drinks[num_drinks['Where do you typically drink coffee?'].str.contains('cafe|go')]

# change how many drinks per day to numberical where 0 is 'less than 1' and 5 is 'more than 4'
num_drinks['How many cups of coffee do you typically drink per day?'] = num_drinks['How many cups of coffee do you typically drink per day?'].apply(coffees_num)

ages_count = num_drinks.groupby(by=['What is your age?']).agg({'How many cups of coffee do you typically drink per day?': 'count'})
ages_count
```

```python
excluded = ['<18 years old', '55-64 years old', '>65 years old']
melted = pd.melt(num_drinks, id_vars=['What is your age?'], value_vars=['How many cups of coffee do you typically drink per day?'])
melted = melted.rename(columns={"What is your age?": "Age"})
melted = melted[~melted['Age'].isin(excluded)]
melted = melted.dropna(subset=['value', 'Age'])


# hist -> https://seaborn.pydata.org/generated/seaborn.histplot.html
plt.figure(figsize=(10, 6))
plt.title('Frequency of number of coffees consumed daily by age')
sns.histplot(melted, x='value', hue='Age', multiple="dodge", shrink=3.5)
```

## Post HOC Analysis

```python
# run posthoc
posthoc = pairwise_tukeyhsd(melted['value'], melted['Age'], alpha=0.05)
print(posthoc)

# create figure
fig = posthoc.plot_simultaneous()
```

## Gender

```python
df = pd.read_csv('data/GACTT_RESULTS_ANONYMIZED_v2.csv', index_col='Submission ID')

# we care about how much coffee people are drinking per day
num_drinks = df.dropna(subset=['How many cups of coffee do you typically drink per day?', 'Where do you typically drink coffee?'])

# we only really want people who drink coffee on the go or at cafes
num_drinks = num_drinks[num_drinks['Where do you typically drink coffee?'].str.contains('cafe|go')]

# change how many drinks per day to numberical where 0 is 'less than 1' and 5 is 'more than 4'
num_drinks['How many cups of coffee do you typically drink per day?'] = num_drinks['How many cups of coffee do you typically drink per day?'].apply(coffees_num)

ages_count = num_drinks.groupby(by=['What is your age?']).agg({'How many cups of coffee do you typically drink per day?': 'count'})
ages_count
```

# Favoured Drinks and Flavours

```python
drinks = df.groupby(by=['What is your favorite coffee drink?']).agg({'What is your age?': 'count'}).rename(columns={'What is your age?': 'count'})
drinks = drinks.sort_values(by='count', ascending=True)

include = ['35-44 years old', '45-54 years old']
drinks_35_54 = df[df['What is your age?'].isin(include)]
drinks_35_54 = drinks_35_54.groupby(by=['What is your favorite coffee drink?']).agg({'What is your age?': 'count'}).rename(columns={'What is your age?': 'count'})
drinks_35_54 = drinks_35_54.sort_values(by='count', ascending=True)

# make plots
plt.figure(figsize=(10, 6))
plt.title('Frequency of prefered drinks')
plt.xticks(rotation=45, ha='right')
sns.barplot(drinks, x='What is your favorite coffee drink?', y='count', label='All ages')
sns.barplot(drinks_35_54, x='What is your favorite coffee drink?', y='count', label='Ages 35-54')
plt.legend()
```

```python
def group_espresso(df):
    # list of drinks that contain espresso
    espresso_drinks = ['Cappuccino', 'Cortado', 'Latte', 'Mocha', 'Americano', 'Espresso']
    
    # create column for identifying espresso drinks
    espresso_id = df
    espresso_id['espresso'] =  espresso_id['What is your favorite coffee drink?'].isin(espresso_drinks)
    espresso_grouped = espresso_id.groupby('espresso').agg({'What is your age?':'count'}).rename(columns={'What is your age?': 'count'})
    espresso_grouped = espresso_grouped[espresso_grouped.index == True]
    count = espresso_grouped.iloc[0, 0]
    
    # remove espresso drinks from drinks df and group remaining drinks 
    no_espresso = espresso_id[espresso_id['espresso'] == False]
    no_espresso = no_espresso.groupby(by=['What is your favorite coffee drink?']).agg({'What is your age?': 'count'}).rename(columns={'What is your age?': 'count'})
    
    # add the new row of espresso drinks
    # create row to add 
    new_row = pd.DataFrame([{"What is your favorite coffee drink?": "Espresso Drink", "count": count}]).set_index('What is your favorite coffee drink?')
    
    # append the new row to the DataFrame
    edrinks = pd.concat([new_row , no_espresso], ignore_index=False)
    edrinks = edrinks.sort_values(by='count', ascending=True)
    return edrinks

# get all ages grouped espresso
espresso_all = group_espresso(df)

# get 35-54 grouped espresso
include = ['35-44 years old', '45-54 years old']
df_35 = df[df['What is your age?'].isin(include)]
espresso_35_54 = group_espresso(df_35)

# make plots
plt.figure(figsize=(10, 6))
plt.title('Frequency of prefered drinks (all espresso drinks grouped)')
plt.xticks(rotation=45, ha='right')
sns.barplot(espresso_all, x='What is your favorite coffee drink?', y='count', label='All ages')
sns.barplot(espresso_35_54, x='What is your favorite coffee drink?', y='count', label='Ages 35-54')
plt.legend()
```
