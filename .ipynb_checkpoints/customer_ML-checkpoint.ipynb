```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import seaborn as sns
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import FunctionTransformer, OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, HistGradientBoostingClassifier
from sklearn.impute import SimpleImputer
import sys

sns.set_theme(style="whitegrid")
```

```python
model = make_pipeline(
    ColumnTransformer([
        # https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html
        ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False), ['age', 'fav_drink', 'Gender', 'taste']),
    ], remainder='passthrough'), 
    GradientBoostingClassifier(
        n_estimators=50,
        max_depth=8,
        min_samples_leaf=5)
)
```

```python
df = pd.read_csv('data/coffee_cleaned.csv', index_col='Submission ID')

df = df.dropna(subset=['most_expensive_purchase_cat', 'age', 'Gender', 'fav_drink', 'taste'])
X = df[['age', 'fav_drink', 'Gender', 'taste', 'whole_milk', 'skim_milk', 'half_and_half', 'creamer', 
        'flav_creamer', 'oat_milk', 'almond_milk', 'soy_milk', 'other_dairy', 'sugar', 'art_sweet', 'honey', 'maple_syrup', 
        'stevia', 'agave_nectar', 'brown_sugar',  'raw_sugar']]
y = df[['most_expensive_purchase_cat']]

X_train, X_valid, y_train, y_valid = train_test_split(X, y)
model.fit(X_train, y_train)

print(model.score(X_train, y_train))
print(model.score(X_valid, y_valid))
```
